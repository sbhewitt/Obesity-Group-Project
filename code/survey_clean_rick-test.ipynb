```python
import pandas as pd
```

```python
df = pd.read_csv('../data/survey.csv')
print(df.shape)
df.head(15)
```

```python
df.isnull().sum()
```

```python
df.rename(columns = {'family_history_with_overweight':'fam_hx_overweight', 
                     'MTRANS':'transport',
                     'FAVC':'high_cal_foods',
                     'FCVC':'vegetables',
                     'NCP':'main_meals',
                     'CAEC': 'food_bw_meals',
                     'SMOKE': 'smoke',
                     'CH2O': 'H2O',
                     'SCC': 'monitor_cals',
                     'FAF': 'physical_activity',
                     'TUE': 'tech_use',
                     'CALC': 'alcohol',
                     'NObeyesdad': 'obesity'}, 
          inplace = True)
df.head()
```

```python
df['fam_hx_overweight'] = df['fam_hx_overweight'].map({'yes':1, 'no':0})
df['high_cal_foods'] = df['high_cal_foods'].map({'yes':1, 'no':0})
df['smoke'] = df['smoke'].map({'yes':1, 'no':0})
df['monitor_cals'] = df['monitor_cals'].map({'yes':1, 'no':0})
```

```python
df = pd.get_dummies(data = df, columns = ['Gender', 'food_bw_meals', 'alcohol', 'transport'], drop_first = True)
```

```python
df['obesity'].value_counts(normalize = True)
```

```python
#Change Values
df['obesity'] = df['obesity'].map({'Obesity_Type_I':3, 'Obesity_Type_II':3, 'Obesity_Type_III':3,
                                         'Overweight_Level_I':2, 'Overweight_Level_II':2, 'Normal_Weight':1, 'Insufficient_Weight':0})
```

```python
df.rename(columns = {'transport_Public_Transportation':'transport_Public'}, inplace = True)
df.head()
```
